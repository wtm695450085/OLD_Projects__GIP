### Krok 1. Tworzenie zbioru danych 

```python
import numpy as np
import matplotlib.pyplot as plt  



import numpy as np
import matplotlib.pyplot as plt
from pylab import *

X = np.linspace(0, 20, num = 50) 
y = 4.45 * np.sin(1.2 * X) + np.random.normal(size = 50)   

plt.figure(figsize=(18,6))

plt.plot(X, y, 'ko', color ='blue', label="Original Noised Data")
plt.legend()
plt.show()  

```

```python
X = np.reshape(X, X.shape + (1,))
X.shape
```

```python
y = np.reshape(y, y.shape + (1,))
y.shape
```

```python
rows_X, col_X = X.shape
rows_y, col_y = y.shape

print("rows_X %s, col_X %s" % X.shape)
print("rows_y %s, col_y %s" % y.shape)
```

```python
# Wymiar wag 
H=30
```

```python
w1 = np.random.randn(col_X, H)
w2 = np.random.randn(H, col_y)

print(w1.shape)
print(w2.shape)
```

```python
learning_rate = 0.00001
for t in range(6000):
  # Forward pass: compute predicted y
  h = X.dot(w1)   #<-- macierz (64 wiersze na 1000 zmiennych) wartości wejściowych jest mnożony przez wektor wag w1 (1000 wierszy na 100 kolumn)
  h_relu = np.maximum(h, 0) #<-- eliminacja ujemnych wartości macierzy H
  y_pred = h_relu.dot(w2)
  
  # Compute and print loss
  loss = np.square(y_pred - y).sum()
  print(t, loss)
  
  # Backprop to compute gradients of w1 and w2 with respect to loss
  grad_y_pred = 2.0 * (y_pred - y)
  grad_w2 = h_relu.T.dot(grad_y_pred)
  grad_h_relu = grad_y_pred.dot(w2.T)
  grad_h = grad_h_relu.copy()
  grad_h[h < 0] = 0
  grad_w1 = X.T.dot(grad_h)
 
  # Update weights
  w1 -= learning_rate * grad_w1
  w2 -= learning_rate * grad_w2
    
  if t % 200 == 0:                             #3. Pokazuje naukę co 100 kroków
       # plot and show learning process
           plt.cla()
           plt.scatter(X, y)
           plt.plot(X, y_pred, 'r--', lw=2)
           plt.text(0.65, 0.1, 'loss=%.4f' % loss, fontdict={'size': 10, 'color':  'black'})
           plt.text(0.65, 0.3, 'step = %d' % t, fontdict={'size': 10, 'color':  'black'})
           plt.pause(0.1)

plt.show()
```
