```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

Tworzę dane

```python
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal');
```

PAC jest uczeniem bez nadzoru czyli analizujemy związek między x i y.
Gdybyśmy robili tutaj uczenie nadzorowane, wprowadzili byśmy regresję liniową określającą y na podstawie wartości x.

W analizie PAC relacja ta jest badana poprzez znalezienie listy głównych osi w danych i użycie tych osi do opisania zestawu danych. 

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_)
```

```python
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=3,
                    color='red',
                    shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# plot data
plt.scatter(X[:, 0], X[:, 1], alpha=0.3)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal');
```

Wektory te reprezentują główne osie danych, a długość wektora wskazuje, jak „ważna” ta oś jest w opisie rozkładu danych - a ściślej, jest to miara wariancji danych podczas projekcji na tej osi. Rzutowanie każdego punktu danych na główne osie jest „głównymi składnikami” danych.

```python
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
pca = PCA(n_components=2, whiten=True)
pca.fit(X)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

# plot data
ax[0].scatter(X[:, 0], X[:, 1], alpha=0.2)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v, ax=ax[0])
ax[0].axis('equal');
ax[0].set(xlabel='x', ylabel='y', title='input')

X_pca = pca.transform(X)
ax[1].scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.2)
draw_vector([0, 0], [0, 3], ax=ax[1])
draw_vector([0, 0], [3, 0], ax=ax[1])
ax[1].axis('equal')
ax[1].set(xlabel='component 1', ylabel='component 2',
          title='principal components',
          xlim=(-5, 5), ylim=(-3, 3.1))
```

## PCA redukuje wymiary
redukcji wymiarowów przez PCA polega na wyzerowaniu jednego lub więcej najmniejszych głównych elementów, co powoduje rzutowanie danych w niższych wymiarach, co pozwala zachować maksymalną wariancję danych.


```python
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)
```

```python
X[:5]
```

```python
X_pca[:5]
```

Komenda n_components=1 oznacza, że zredukowaliśmy zmienne do jednej zmiennej, jednego wymairu.

```python
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8)
plt.axis('equal');
```

Jasne punkty to oryginalne dane, a ciemne punkty to wersja rzutowana.
PCA: punkty (x,y) wzdłuż najmniej ważnej głównej osi lub wielu osi są usuwane, pozostawiając tylko (x,y) o największej wariancji.

Część punktów (x,y) o małej wariancji, które są wycinane (proporcjonalna do rozrzutu punktów wokół linii utworzonej na rysunku), jest miarą tego, ile „informacji” jest odrzucanych przy redukcji wymiarów.

Ten zestaw danych ze zredukowanym wymiarem jest „wystarczająco dobry”, aby zakodować najważniejsze relacje między punktami (x,y): pomimo zmniejszenia wymiaru danych o 50%, ogólna relacja między punktami danych jest w większości zachowana.
