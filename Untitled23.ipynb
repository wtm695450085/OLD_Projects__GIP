```python
dff = df[df.countriesAndTerritories=='United_States_of_Americ']
dff = dff.loc['2020-04-01':'2020-04-30'] 
dff.sort_index(ascending=True)
```

```python
import matplotlib.pyplot as plt

dff[['cases','deaths']].plot(figsize=(20,10), linewidth=1, fontsize=20)
plt.xlabel('Year', fontsize=20)
```

```python
Kolumny = ['cases', 'deaths']
df_weekly = dff[Kolumny].resample('W').sum()
df_weekly.head(20)
```

```python
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

sns.set(rc={'figure.figsize':(17, 6)})

start, end = '2020-01', '2020-04'

fig, ax = plt.subplots()
#ax.plot(df.loc[start:end, 'cases'],
#marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(df_weekly.loc[start:end, 'cases'],
marker='o', markersize=9, linestyle='-',linewidth=1.5, label='Weekly Mean Resample')
ax.set_ylabel('Solar Production (GWh)')
ax.legend();
```
